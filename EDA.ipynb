```python
#Importing required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Importing the cleaned dataset

df = pd.read_csv('cleaned_mall_customers.csv')
print(df.head(10))
```

```python
#Descriptive Statistics

#Mean, Median and Mode

for col in ['Age', 'Annual Income', 'Spending Score']:
    print(f"Mean {col} : {df[col].mean()}")
    print(f"Median {col} : {df[col].median()}")
    print("\n")

print("Mode of Gender: ",df['Gender'].mode()[0])
```

```python
#Variance and Standard Deviation

for col in ['Age', 'Annual Income', 'Spending Score']:
    print(f"Standard Deviation in {col} : {df[col].std()}")
    print(f"Variance in {col} : {df[col].var()}")
    print("\n")
```

```python
#Data Visualization
```

```python
#Histplots showing Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

#Histplots showing Annual Income Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Annual Income'], bins=30, kde=True)
plt.title('Annual Income Distribution')
plt.show()

#Histplots showing Spending Score Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Spending Score'], bins=30, kde=True)
plt.title('Spending Score Distribution')
plt.show()
```

```python
# Visualizing relationship between Spending Score and Annual Income 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income', y='Spending Score', hue='Gender')
plt.title('Income vs Spending Score')
plt.show()
```
